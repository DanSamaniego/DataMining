### (a) Download the Echocardiogram data set and import it to a data frame. Read the information on the data set and find out how missing values are represented and make sure that they are properly represented.

```python
# Descargar el conjunto de datos de ecocardiograma e importarlo a un marco de datos.
# Lea la información en el conjunto de datos y descubra cómo se representan 
# los valores faltantes y asegúrese de que estén representados correctamente.
import csv
import IPython

datos = []

# Leer el archivo CSV original
with open('/kaggle/input/echocardiagram/echocardiogram.data', 'r') as archivoEcho_csv:
    lectorEcho_csv = csv.reader(archivoEcho_csv)
    datos = [fila for fila in lectorEcho_csv]
    
# Recorre los datos y reemplazar "?" por "NA" para asegurarnos que los datos estén
# representados correctamente.
for fila in datos:
    for i in range(len(fila)):
        if fila[i] == "?":
            fila[i] = "NA"

# Guarda el archivo actualizado
with open('/kaggle/working/echocardiogram-updated.csv', 'w', newline='') as archivoEcho_actualizado:
    escritorEcho_csv = csv.writer(archivoEcho_actualizado)
    escritorEcho_csv.writerows(datos)
    
# Lee el archivo CSV actualizado
df_echoUpdate = pd.read_csv('/kaggle/working/echocardiogram-updated.csv')

# Imprime el DataFrame actualizado
print("Archivo actualizado guardado correctamente.\n")
print(df_echoUpdate)
```

### (s) Create a new attribute TempC which represents the temperature values in Celsius.

```python
# Lee el archivo CSV
df = pd.read_csv('/kaggle/working/formatoTrama.csv')

# Aplicar la formula de conversion para crear el nuevo atributo tempC
# C = (F - 32) * 5/9
df['tempC'] = (df['Temp'] - 32) * 5/9

# Redondea el atributo tempC a 2 decimales
df['tempC'] = round(df['tempC'], 2)

# Imprime el DataFrame con el nuevo atributo
print(df)
```

### (t) Inspect which were the 30 hottest days

```python
# Ordenar el DataFrame por el atributo tempC de dorma descendente
df_Descendente = df.sort_values('tempC', ascending=False)

# Obtenemos los 30 dias mas calurosos
diasCalurosos = df_Descendente.head(30)

# Imprime los 30 dias mas calurosos
print(diasCalurosos)
```

### (u) Inspect which were the hottest days, but also with the highest ozone values.

```python
# Obtener los 30 dias mas calurosos
diasCalurosos = df.nlargest(30, 'tempC')
# nalargest sirve para encontrar los N valores mas grandes (Orden Descendente)
# en una serie o columna de un Dataframe 

# Obtener los 30 dias con los valores de Ozono mas altos
diasAltoOzono = df.nlargest(30, 'Ozone')

# Combina los resultados
diasCalorOzono = diasCalurosos.merge(diasAltoOzono)

# Imprime los dias mas Calurosos con los valores mas Altos de Ozono
print(diasCalorOzono)
```

### (v) Inspect the number of days for which there was a register for each month.

```python
# Contar el número de registros por mes
registrosPorMes = df.groupby('Month').size() # Size cuenta el numero de registros por mes
#registros_por_mes = df.groupby('Month').size()

# Filtrar los meses con un solo registro
mesesUnRegistro = registrosPorMes[registrosPorMes == 1].index
#meses_con_un_registro = registros_por_mes[registros_por_mes == 1].index

# Imprimir los meses con un solo registro
print('Meses con un solo registro')
for mes in mesesUnRegistro:
    print('Mes', mes)
    
print('Todos los meses tienen mas de un registro por lo que no imprime nada')
```

### (w) For each month, obtain the minimum and the maximum temperature registered in Celsius.

```python
# Obtener la temperatura minima y maxima por Mes
tempMes = df.groupby('Month')['tempC'].agg(['min', 'max'])
# agg calcula la temperatura minima y maxima para cada grupo

# Imprimir los resultados
for mes, temperatura in tempMes.iterrows():
    minima = temperatura['min']
    maxima = temperatura['max']
    print(f'Mes {mes}: Temperatura minima = {minima:.2f}°C, Temperatura maxima = {maxima:.2f}°C')
```

### (x) Obtain the average of the following parameters by month: temperature in celsius, wind, solar radiation and ozone.

```python
# Calcular el promedio por mes de temperature in celsius, wind, solar radiation y ozone
promedioTemperatura = df.groupby('Month')['tempC'].mean()
promedioViento = df.groupby('Month')['Wind'].mean()
promedioRadiacion = df.groupby('Month')['Solar.R'].mean()
promedioOzono = df.groupby('Month')['Ozone'].mean()

# Imprime los resultados
print('Promedio por mes de Temperatura en grados Celsius:', promedioTemperatura)

print('\nPromedio por mes del Viento:', promedioViento)

print('\nPromedio por mes de Radiacion:', promedioRadiacion)

print('\nPromedio por mes de Ozono:', promedioOzono)
```
