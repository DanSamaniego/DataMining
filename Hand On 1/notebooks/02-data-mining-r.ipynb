### (p) What period of the year do these records refer to?

```r
# Extrae el periodo del año
datosAq$Periodo <- ifelse(datosAq$Month %in% c(3, 4, 5), 'Primavera',
                         ifelse(datosAq$Month %in% c(6, 7, 8), 'Verano',
                               ifelse(datosAq$Month %in% c(9, 10, 11), 'Otoño',
                                    'Invierno')))

# Imprime los registros con la informacion del periodo
print(datosAq)
```

### (q) Load the package dplyr and save the data set in a table data frame format.

```r
# Cragar el paquete dplyr
library(dplyr)

# Convierte el conjunto de datos en una tabla de datos manipulable
tabla_datos <- as_tibble(datosAq)

# Imprime la tabla de datos
print(tabla_datos)
```

### (r) Select the days in May with a temperature above 70 Fahrenheit.

```r
# Carga el paquete dplyr
library (dplyr)

# Filtrar los dias de Mayo con una temperatura superior a 70 grados Fahrenheit
diasMayoCalurosos <- tabla_datos %>%
    filter(Month == 5, Temp > 70)

# Imprime los resultados
print(diasMayoCalurosos)
```

### (s) Create a new attribute TempC which represents the temperature values in Celsius.

```r
# Cragar el paquete dplyr
library (dplyr)

# Convertir la temperatura de Fahrenheit a Celsius y guaradar en un nuevo atributo
tabla_datos <- tabla_datos %>% mutate(TempCelsius = (as.numeric(Temp) - 32) * 5/9)

# Imprrmir los datos actualizados
print(tabla_datos)
```

### (t) Inspect which were the 30 hottest days

```r
# Ordenar los datos por temperatura de forma descendente 
diasCalurosos <- tabla_datos %>%
    arrange(desc(Temp))

# Seleccionar los 30 dias mas calurosos
diasCalurosos <- head(diasCalurosos, 30)

# Mostrar la tabla con los 30 dias mas calurosos
View(diasCalurosos)
```

### (u) Inspect which were the hottest days, but also with the highest ozone values.

```r
# Dias mas calurosos con los valores mas altos de ozono
library(dplyr)

# top.n Selecciona los 30 registros con temperaturas mas altas
resultados <- tabla_datos %>%
    top_n(30, Temp) %>%
    arrange(desc(Temp)) %>%  # Ordena en funcion de la temperatura de forma descendente y luego por el Ozono de forma descendente
    arrange(desc(Ozone))

# Imprime la consulta
print(resultados)
```

### (v) Inspect the number of days for which there was a register for each month.

```r
# Agrupa el conjunto de datos por mes
mesesUnRegistro <- tabla_datos %>%
group_by(Month) %>%

# Aplica filtro para mantener los meses que tienen un unico registro
filter(n() == 1) %>%

# Obtiene los meses unicos
distinct(Month)

# Imprime los meses que tienen un registro
print('Meses que tienen un solo registro')
print(mesesUnRegistro)
```

### (w) For each month, obtain the minimum and the maximum temperature registered in Celsius.

```r
library (dplyr)

# Agrupar por mes y obtener la temperatura maxima y minima
datosTempMeses <- tabla_datos %>%
group_by(Month) %>%
# summarise toma como argumentos las columnas de la tabla a resumir y las funciones
# de resumen que se aplicaran a cada columna
summarise(Temperatura_Min = min(TempCelsius, na.rm = TRUE),
         Temperatura_Max = max(TempCelsius, na.rm = TRUE))

# Imprime los resultados
print(datosTempMeses)
```

### (x) Obtain the average of the following parameters by month: temperature in celsius, wind, solar radiation and ozone.

```r
# Agrupar los datos por mes y calcular los promedios
library(dplyr)

promedios <- tabla_datos %>%
    group_by(Month) %>%       # Agrupa por mes 
# summarise toma como argumentos las columnas de la tabla a resumir y las funciones
# de resumen que se aplicaran a cada columna
    summarise(Promedio_Temp_Celsius = mean(TempCelsius, na.rm = TRUE),
             Promedio_Viento = mean(Wind, na.rm = TRUE),
             Promedio_Rad_Solar = mean(ifelse(is.numeric(Solar.R), Solar.R, NA), na.rm = TRUE),
             Promedio_Ozono = mean(ifelse(is.numeric(Ozone), Ozone, NA), na.rm = TRUE)) %>%
    ungroup()
# na.rm = TRUE   indica que se deben omitir los valores faltantes (NA) al calcular el resultado

# Imprime los promedios por mes
print('Los promedios por mes son:')
print(promedios)
```

### (y) What values did you obtain regarding ozone and solar radiation attributes? Why? Make the necessary change so that you get the average of the registered values.

```r
library(dplyr)

# Suprime los mensajes de advertencia
suppressWarnings({
  # Convierte las columnas a tipo numérico
  tabla_datos$Solar.R <- as.numeric(tabla_datos$Solar.R)
  tabla_datos$Ozone <- as.numeric(tabla_datos$Ozone)
  
  # Elimina filas con NA en las columnas Solar.R y Ozone
  tablaDatosFiltrado <- tabla_datos[complete.cases(tabla_datos$Solar.R, tabla_datos$Ozone), ]
  
  print('Tabla actualizada:')
  print(tablaDatosFiltrado)
})

# Agrupar los datos por mes y calcular los promedios
promedios <- tablaDatosFiltrado %>%
    group_by(Month) %>%       # Agrupa por mes 
# summarise toma como argumentos las columnas de la tabla a resumir y las funciones
# de resumen que se aplicaran a cada columna
    summarise(Promedio_Temp_Celsius = mean(TempCelsius, na.rm = TRUE),
             Promedio_Viento = mean(Wind, na.rm = TRUE),
             Promedio_Rad_Solar = mean(Solar.R, na.rm = TRUE),
             Promedio_Ozono = mean(Ozone, na.rm = TRUE)) %>%
    ungroup()
# na.rm = TRUE   indica que se deben omitir los valores faltantes (NA) al calcular el resultado

# Imprime los promedios por mes
print('Los promedios por mes son:')
print(promedios)
```
