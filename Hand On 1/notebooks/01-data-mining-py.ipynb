### (a) Create an array of 10 random integers between 1 and 100, and then calculate the mean and standard deviation of the array.

```python
# Create an array of 10 random integers between 1 and 100
# Estamos creando un array con 10 números NO ALEATORIOS
a = [25, 16, 74, 96, 35, 41, 63, 17, 3, 65]

# Calcula la cantidad de elementos del array y los imprime
print('Number of elements: ', len(a), '\nArray: ', a)    

# Indica el tipo de dato que se le pasa como argumento
print('Type: ', type(a))

# Calcula e imprime la Media Aritmetica del array a
print('Mean: ',np.mean(a))      

# Calcula e imprime la Desviacion Estandar del arrar a
des_est = np.std(a)
print('Standar Desviation: ', des_est)       # Desviacion estandar

# Al ser array tambien se utiliza para texto y al hacer mean no sería posible como lo es
# en el caso b que es un array np y no solo funciona con números
```

```python
# Create an array of 10 random integers between 1 and 100
b = np.random.randint(1, high=101, size=10)

# Calcula la cantidad de elementos del array y los imprime
print('Number of elements: ',len(b), '\nArray: ', b)

# Calcula el tipo de tipo de dato que se le pasa como argumento
print('Type: ', type(b))

# Calcula e imprime la Media Aritmetica del array
mean = b.mean()
print("Mean:", mean)

# Calcula e imprime la Desviacion Estandar del array
# print('Desviacion Estandar: ', b.std()) Dos formas de calcular la desviacion estandar
print('Standar Desviation: ', np.std(b))
```

## (b) Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column

```python
# Create a 2-dimensional array of 3 rows and 4 columns with random integer values. 
mi_matriz = np.random.randint(1, 10, size=(3, 4)) # Numeros aleatorios del 1 al 9

# Print the matrix with random values
print ("Original matrix: \n", mi_matriz )

# Calculate the sum of each column.
sumCols = np.sum(mi_matriz, axis=0)
print ("\nSum of each column: ", sumCols)

# Calculate the sum of each row
sumRows = np.sum(mi_matriz, axis=1)
print ("\nSum of each row:", sumRows)
# axis indica la dimension de la matriz
```

## (c) Create a DataFrame with 3 columns: "Name", "Age", and "City". Add at least 5 rows of data to the DataFrame. Then, filter the DataFrame to only include rows where the person's age is greater than or equal to 30.

```python
# Create a DataFrame with 3 columns: "Name", "Age", and "City".
df = pd.DataFrame()
df['Name'] = None
df['Age'] = None
df['City'] = None

# Print the empty Data Frame
print("Data Frame vacio\n")
print(df)

# Add at least 5 rows of data to the Data Frame
# En forma de listas
names = ["Luis","Pedro","Camila","Elizabeth","Rous"]
ages = [24,15,41,36,12]
cities = ["Otavalo","Quito","Ibarra","Cuenca","Chaco"]

df["Name"] = names
df["Age"] = ages
df["City"] = cities

# Llenado con un objeto Series
##newRow = pd.Series(["Juan",18,"Manta"], index=df.columns) # Index coge las columnas del df escritas antes para no escribirlas manualmente
# El método append() añade un elemento al final de la lista
##df = df.append(newRow, ignore_index=True) # Ignora el índice cuando por ej los datos vienen de otro dataframe que sí tiene índice

# Llenado con un diccionario
##newRow = {"Name": "Blanca", "Age": 52, "City": "Loja"}
##df = df.append(newRow, ignore_index=True)

# Imprime el Data Frame lleno con los elementos insertados
print("\nFull data frame:\n", df)

# Filter, only include rows where the person's age is greater than or equal to 30
# Filter the DataFrame
greaterEqual = df["Age"] >= 30
dfFilter = df[greaterEqual]

# Imprecion de Data Frame aplicado el filtro.
print("\nData frame with filter:\n", dfFilter)
```

## (d) Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the columns in the DataFrame.

```python
# Load a CSV file into a DataFrame
serviceArea = pd.read_csv('/kaggle/input/servicearea/ServiceArea.csv')

# Print the read file
print(serviceArea)

# Calculate and print the mean, median, and mode of one of the columns in the DataFrame.
print("\nMean: ", serviceArea["VersionNum"].mean())
print("Median: ", serviceArea["VersionNum"].median())
print("Mode: ", serviceArea["VersionNum"].mode())
```

## Create a scatter plot of random x and y values between 1 and 100.

```python
import matplotlib.pyplot as plt

# Generate random x and y values
x = np.random.randint(low=1, high=100, size=20)
y = np.random.randint(low=1, high=100, size=20)

# Create a scatter plot and add color
plt.scatter(x, y, color="red")

# Add title to the graph
plt.title('Scatter Plot - X and Y Values')

# Add title to the axes
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Show the graph
plt.show()
```
