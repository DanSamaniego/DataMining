### (a) Create an array of 10 random integers between 1 and 100, and then calculate the mean and standard deviation of the array.

```r
# Create an array of 10 intergers between 1 and 100
a <- c(25, 16, 74, 96, 35, 41, 63, 17, 3, 65)
a <- array(a)

# Print array
print('Array:')
a

# Calculate the mean
#print(paste('Mean: ', mean(a)))  # Otra forma de imprimir
x <- mean(a)

# Calculate the standard deviation
y <- sd(a)   # averiguar por qué sale diferente que lo de python

# Print the mean and standard deviation
print(paste('Mean: ',x))
print(paste('Standard Deviation: ', y))
```

```r
# Create an array of 10 random integers between 1 and 100
b <- runif(10, min=1, max=100)    # Decimales aleatorios

# Print decimal array
print('Decimal Array:')
b

# Create an array of 10 random integers between 1 and 100
c <- sample(1:100, 10)   # Enteros aleatorios

# Print array of integers
print('Integers Array:')
c

# Calculate the mean and standard deviation
x <- mean(c)   # Media aritmética
y <- sd(c)     # Desviacion Estandar

# Print the mean and standard deviation
print(paste('Mean:', x))
print(paste('Standard Deviation: ', y))
```

## (b) Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column

```r
# Create a 2-dimensional array of 3 rows and 4 columns with random integer values. 
rmatriz <- matrix(sample.int(10, size=12, replace=TRUE), nrow=3, ncol=4)
rmatriz

# Calculate and print the sum of each row
print("Suma de cada Fila:")
rowSums(rmatriz)

# Calculate and print the sum of each column
print("Suma de cada Columna:")
colSums(rmatriz)

#sprintf('Suma filas es %s', rowSums(rmatriz)) # Otra forma de imprimir
```

## (c) Create a DataFrame with 3 columns: "Name", "Age", and "City". Add at least 5 rows of data to the DataFrame. Then, filter the DataFrame to only include rows where the person's age is greater than or equal to 30.

```r
# Create a DataFrame with 3 columns: "Name", "Age", and "City". 
# Add at least 5 rows of data to the DataFrame.
names <- c('Mauricio', 'Camilo', 'Adriana', 'Joss', 'Viviana')
ages <- c(45, 12, 26, 74, 34)
cities <- c('Cuenca', 'Pillaro', 'Latacunga', 'Quito', 'Ibarra')

# Print Data Frame
datos <- data.frame(Name=names, Age=ages, City=cities)
datos

# Adding a new record to the data frame (Agregar un nuevo registro al data frame)
newReg <- data.frame(Name='Pedro', Age=15, City='Otavalo')
datos <- rbind(datos, newReg)    # une agregando filas, con las columnas iguales

# Prints updated data frame
'Data Frame con datos actualizados'
datos

# Filter the DataFrame to only include rows where the person's age is greater than or equal to 30
personfilter <- subset(datos, Age>=30)

# Print data frame with filtered people
'Data Frame con edades filtradas'
personfilter

# Print data frame with filtered people
'Otra forma de filtrar edades'
datos[datos$Age >= 30,] # Otra forma de filtrar por edad
```

## (d) Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the columns in the DataFrame.

## (e) Create a scatter plot of random x and y values between 1 and 100.

```r
# # Generate random x and y values
x <- runif(30,1, 50) # 30 numeros aleatorios entre 1 y 50
y <- runif(30,1, 50)

# Agregamos la funcion PLOT para graficar, añadimos título, nombre de los ejes 
# PCH que nos permite modificar el símbolo de los puntos en el grafico y agregamos color
plot(x, y, main='Scatter plot X vs Y', xlab='X', ylab='Y', pch=25, col='orange')
```
